## Setup

`````python
# Using Google Gemini (free tier) instead of OpenAI.
# Put your key in a .env file in this folder as:   GOOGLE_API_KEY=your-key-here
# Get a free key at: https://aistudio.google.com/app/apikey

from google import genai
from google.genai import types
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())  # read local .env file

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
`````

`````python
import time
from google.genai import errors as genai_errors

client = genai.Client(api_key=GOOGLE_API_KEY)

def get_completion(prompt, model="gemini-2.5-flash", max_retries=8):
    """
    Drop-in replacement for the OpenAI helper, backed by Gemini.
    Retries on 429 (free-tier rate limit) with exponential backoff + small
    baseline sleep to smooth out bursts.
    """
    time.sleep(4)  # gentle pacing to stay under free-tier per-minute limits
    for attempt in range(max_retries):
        try:
            response = client.models.generate_content(
                model=model,
                contents=prompt,
                config=types.GenerateContentConfig(
                    temperature=0,  # deterministic output
                ),
            )
            return response.text
        except genai_errors.ClientError as e:
            if getattr(e, "status_code", None) == 429 and attempt < max_retries - 1:
                wait = min(2 ** attempt * 5, 60)
                print(f"Rate limited, waiting {wait}s...")
                time.sleep(wait)
                continue
            raise
`````

## Exercise — My Work

Three summarization experiments that go beyond the simple "shorten a review" pattern:

1. **Audience-aware summarization** — one article, three very different readers.
2. **Progressive compression ladder** — the same text at many length budgets (from 60 words down to a single emoji).
3. **Structured summary (JSON)** — turning a long review into machine-readable fields, not prose.

### Variation 1 — Audience-aware summaries

Goal: show that a good prompt changes *what* counts as relevant. I take a single fictional news article and ask the model to summarize it three different ways — for a CEO, for a software engineer, and for an investor — keeping each summary under 40 words. A single raw text, three very different extracts.

`````python
article = """
Fictional article — BrightPath AI, a Madrid-based startup, announced a 28 million
euro Series B funding round led by Accel and Seaya Ventures. The company builds
an internal-documentation assistant for enterprise engineering teams and says
the fresh capital will be used to (a) triple its engineering team in Madrid and
Lisbon, (b) migrate its vector search infrastructure from pgvector to a custom
Rust-based engine for 3x lower latency, and (c) open a second office in Berlin.
CEO Marta Pons said the company's annual recurring revenue has grown 4x in the
last twelve months and that customer renewal sits at 93%. Competitors Glean
and Dust have raised at higher valuations but BrightPath claims a shorter
integration time (under 48 hours for a typical mid-size customer).
"""

audiences = {
    "CEO of a competing company": "strategic threat and recommended response",
    "Software engineer evaluating it for their team": "technical architecture and integration effort",
    "Early-stage VC investor": "growth metrics, valuation signals, and market positioning",
}

for audience, focus in audiences.items():
    prompt = f"""
    Summarize the article below for the following reader.
    Reader: {audience}
    What the reader cares about: {focus}

    Constraints:
    - Under 40 words.
    - Only include information that is in the article.
    - Do NOT add disclaimers or opinions.

    Article: ```{article}```
    """
    print(f"--- For: {audience} ---")
    print(get_completion(prompt))
    print()
`````

### Variation 2 — Progressive compression ladder

Goal: take the same product review and force the model down a ladder of length budgets — 60 words, 20 words, 5 words, and finally a single emoji — to observe how meaning degrades as the budget shrinks.

`````python
text_to_compress = reviews[1]  # the standing-lamp review

budgets = ["exactly 60 words", "exactly 20 words", "exactly 5 words", "a SINGLE emoji (no text)"]

for b in budgets:
    prompt = f"""
    Summarize the review below in {b}.
    Preserve the main sentiment if possible.
    Output ONLY the summary, no explanations.

    Review: ```{text_to_compress}```
    """
    print(f"[{b}] -> {get_completion(prompt)}")
`````

### Variation 3 — Structured JSON summary

Goal: replace prose with machine-readable fields. Many real-world pipelines do not need "a nice paragraph" — they need a row in a table. I ask the model to emit strict JSON with `pros`, `cons`, `target_buyer`, `recommendation_score` (1-10), and `one_line_verdict`.

`````python
import json, re

def safe_json(text):
    cleaned = re.sub(r"^```(?:json)?\s*|\s*```$", "", text.strip(), flags=re.MULTILINE)
    return json.loads(cleaned)

text_to_structure = reviews[3]  # the blender review

prompt = f"""
Analyze the product review delimited by triple backticks and output ONLY a
JSON object (no prose, no markdown fences) with the following keys:

- "pros": list of short strings (max 3)
- "cons": list of short strings (max 3)
- "target_buyer": one short phrase describing who would most benefit from
  this product based on the review
- "recommendation_score": integer 1-10 (10 = strongly recommend)
- "one_line_verdict": a single sentence, under 25 words

Review: ```{text_to_structure}```
"""

raw = get_completion(prompt)
print("Raw output:\n", raw)

parsed = safe_json(raw)
print("\nParsed:")
for k, v in parsed.items():
    print(f"  {k}: {v}")
`````

## One-page report — findings & learnings

### What I built
1. **Audience-aware summaries** — one BrightPath article summarized three ways (competitor CEO, engineer, VC), each under 40 words.
2. **Progressive compression ladder** — a standing-lamp review compressed at 60 words → 20 words → 5 words → 1 emoji.
3. **Structured JSON summary** — a noisy blender review turned into `pros`, `cons`, `target_buyer`, `recommendation_score`, and `one_line_verdict`.

### What worked
- **Audience framing changes what gets kept.** The CEO summary emphasised integration speed and Madrid hiring; the engineer summary focused on Rust/pgvector and latency; the investor summary led with the 4x ARR and 93% retention. Same article, three materially different extractions.
- **Short length budgets force real compression.** At 20 words the model dropped adjectives and kept facts. At 5 words it kept a single headline idea ("Great lamp, excellent customer service"). The emoji step almost always produces a single relevant emoji (💡 or 🛋️ for the lamp).
- **Structured JSON is more reusable than prose.** The JSON output in variation 3 is directly loadable with `json.loads`, making it easy to feed into a dashboard or a table — prose summaries aren't.
- **Telling the model "only include information from the text"** is the cheapest way to reduce drift: in early tests the competitor-CEO summary wanted to add "you should offer a free trial", which is not in the article; adding the constraint stopped it.

### Where things went wrong / surprises
- **"At most 30 words" is honoured softly.** Gemini and GPT-3.5 both routinely go 2-4 words over. If a hard limit matters (e.g. tweet length), post-validate and truncate.
- **Extreme compression invents meaning.** At the 5-word budget the model sometimes picks a positive framing even for mixed reviews. At the emoji step, it occasionally outputs 2–3 emojis despite the "single emoji" rule. Strict formats need explicit rules AND verification.
- **Focus prompts leak.** The shipping-focused summary from the class still mentions "cute" and "daughter's birthday" despite the instruction to focus on shipping. That's why the notebook itself switched from "summarize" to "extract" — "extract" frames the task as retrieval rather than rewriting, and it is noticeably tighter.
- **Structured summarization fails loudly, which is a good thing.** If the model adds a ```` ```json ```` fence, `json.loads` will throw — that is better than silently returning bad data. The `safe_json` helper strips the fence.
- **Summary quality depends on what you *ask for*, not just what the text says.** Two well-phrased prompts on the same text produce summaries that look almost unrelated. The prompt is part of the summary.

### What I learned
1. **"Summarize" and "extract" are different tasks.** Use "extract" when you want only task-relevant sentences; use "summarize" when you want a gloss of the whole thing.
2. **Always state the audience and the constraint.** `who-is-reading-this` + `how-long` + `what-to-include` + `what-to-exclude` is a better template than a generic "summarize this".
3. **Structured output beats prose for pipelines.** Prompting for JSON + `json.loads` turns a summarization model into a tiny, free ETL step.
4. **Length budgets are soft, not hard.** Never trust the model to count words — validate afterwards.
5. **Free-form output can hallucinate; extractive prompts don't.** Instructing the model to quote directly or to say "NOT IN TEXT" eliminates a whole class of summarization errors.
